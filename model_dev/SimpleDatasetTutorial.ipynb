```python
from settings import Settings
from utils.datasets import MaestroDataset
import matplotlib.pyplot as plt
```

```python
settings = Settings() # Set Extra Settings in settings.py
```

```python
dataset = MaestroDataset("train", settings=settings)
```

```python
print('total number : {}'.format(len(dataset)))
```

```python
def show_image(index):
    piano_roll, velocity = dataset[index]
    
    fig=plt.figure(figsize=(20, 8))
    rows = 1
    columns = 2
    
    fig.add_subplot(rows, columns, 1)
    plt.imshow(piano_roll)
    
    fig.add_subplot(rows, columns, 2)
    plt.imshow(velocity)
    
    plt.show()
```

```python
index = 0
while index!=-1:
    index = int(input())
    show_image(index)
```
